```python
data = pd.read_csv("historical_data.csv")
```

```python
data_from = pd.to_datetime(data['created_at'])
data_to = pd.to_datetime(data['actual_delivery_time'])
```

```python
d = {}

for i in data['order_protocol']:
    if i == np.nan:
        continue
    if i in d:
        d[i] += 1
    else:
        d[i] = 1

for i in d:
    d[i] = [d[i], (d[i]/len(data)) * 100]
print(len(d), data['order_protocol'].unique())
d

# 1,2,3,4,5
```

```python

len(data['store_primary_category'].unique())

d = {}

for i in data['store_primary_category']:
    if i in d:
        d[i] += 1
    else:
        d[i] = 1

for i in d:
    d[i] = [d[i], (d[i]/len(data)) * 100]
print(len(d))
d

#american, mexican
```
